**Importing libraries tensorflow,pandas and Matplotlib**

```python
import tensorflow as tf
from tensorflow import keras
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_theme(style='whitegrid', context = 'paper', font = 'Georgia')
```

**Loading the dataset**

```python
(X_train,Y_train),(x_test,y_test) = keras.datasets.cifar100.load_data(label_mode="fine")
```

**Getting General idea of datasets**

```python
print(f'The shape of X_train is :{X_train.shape}','\t',f'The shape of Y_train is :{Y_train.shape}','\n',
     f'The shape of X_test is :{x_test.shape}','\t',f'The shape of y_test is :{y_test.shape}')
```

**Scaling the datasets**

```python
X_train_scaled = X_train/255
x_test_scaled = x_test/255
```

**data Visualization**

```python
for i in range(12):
    plt.subplot(3,4,i+1)
    plt.imshow(X_train[i])
    plt.axis('off')
```

**Building simple ANN**

```python
model = keras.Sequential([
    keras.layers.Flatten(input_shape = (32,32,3)),
    keras.layers.Dense(3000,activation = 'relu'),
    keras.layers.Dense(3000,activation = 'tanh'),
    keras.layers.Dense(3000,activation = 'sigmoid')

])

model.compile(
    optimizer = 'adam',
    loss = 'sparse_categorical_crossentropy',
    metrics = ['accuracy']
)

model.fit(X_train_scaled,Y_train, epochs = 1)
```
